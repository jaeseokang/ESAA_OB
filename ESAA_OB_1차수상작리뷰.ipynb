### 데이터

- Glucose: 포도당 농도, BloodPressure: 혈압, SkinThickness: 피부 두께, Insulin: 인슐린, BMI: 체질량지수, DiabetesPedigreeFunction: 당뇨병 혈통 기능, Age: 나이, Outcome: 당뇨병 여부(0: 발병x, 1: 발병)

### 목적 
- 당뇨병 위험을 **분류** 모델로 예측

# [코드 리뷰]

### (1) EDA
- train/test set info 확인
- isnull().sum() = 0 값 결측치 처리
- 히트맵 시각화
- 박스플롯으로 이상치 확인 : Data(Glucose, BloodPressur, SkinThickness, Insulin, BMI) -> 이상치 존재 o

### (2) 기능 엔지니어링 및 데이터 정리
- 이상치 처리: Insulin=0은 결측치 처리x, 500 이상 이상치 처리,
- BMI>=50, BloodPressur=0, Glucose=0, SkinThickness>=90 이상치 처리
  
### (3) 모델 훈련 및 생성
- feature/target 분리 (X: ID, Outcome 에 대해)
- 데이터 분할
- Voting model 사용(Random Forest, XGBClassifier, SVC)
- Pseudolabeling/테스트 데이터로 예측한 결과 훈련 데이터에 추가
- 훈련 데이터 확장 -> 다시 모델 학습
- Stacking 모델 구성(Random Forest, XGBClassifier, SVC)
- Thresfold 튜닝
- F1-score 임계값 찾기 -> 최적 임계값 이용해 최종 예측
- accuracy, ROC-AUC, F1 socre 으로 최종 성능 평가

### (4) 후처리
- Validation set까지 포함된 전체 데이터로 모델 재학습해주고 마무리
- 최종 테스트 데이터 예측 및 저장

# [차별 점 및 배울 점]

### [차별 점]
- 여러 교란 변수들에 대해 각자의 기준을 가지고 이상치 설정
- DT 등의 모델 사용 후, 성능이 좋지 않아 Voting model 사용

### [배울 점]
- 먼저 박스플롯으로 이상치 확인 후, 이상치 기준을 세워 처리
- 훈련 데이터를 확장시켜 모델을 재학습 시킴
- 분류 성능 평가로 accuracy 외에도 ROC-AUC, F1 score 등 복합적으로 평가

# [주제 및 데이터]
## 주제 : 식음업장 메뉴 수요 예측 AI 오프라인 해커톤

Dacon 수상작 리뷰 https://dacon.io/competitions/official/236594/overview/description

메타데이트롤 포함한 LSTM 기반 메뉴 매출 수량 예측

## 데이터

1. train.csv(영업장명_메뉴명별 매출 수량 정보, 영업일자, 영업장명_메뉴명, 매출수량)
2. price.csv(영업장명_메뉴명별 평균 판매금액, 영업장명 메뉴명, 평균판매금액)
3. room_type.csv(리조트내 객실 정보, 객실타입, 객실타입명, 기준인원)
4. Map.jpg(리조트내 가이드맵 이미지)
5. test(폴더)
6. sample_submission.csv(파일): 제출양식

# [코드 리뷰]

(1) import
- randomseed & hyperparameter 및 data load(parse_dates=["영업일자"])

(2) 기능 엔지니어링 및 데이터 정리

- 메타 데이터 IO/병합
- 영업일자 날짜만 남겨 일(day) 단위로 조인되게 정규화
- 날짜 기반 메타 병합

(3) 모델 훈련 및 생성

- 모델 정의(LSTM) -> 시계열 모델
- 모델 훈련 / 학습 / 예측
- price 병합
- 스케일 병합 / 역변환

(4) 후처리
- submission.csv 파일로 제출

# [차별 점 및 배울 점]

## [차별 점]

- 메타데이터를 포함해 단순히 과거 매출/수량 등 시계열 데이터만 쓰는게 아니라, 메뉴 특성, 점포 특성 등 다양한 부가정보를 같이 활용
- 메뉴별 매출 수량(quantity) 예측
- error 지표 / loss 함수 활용 많음

## [배울 점]

- baseline 코드로, 기본적이면서도 확장 가능한 구조를 가짐
- 메타데이터가 있어 시간 처리 및 시계열 전처리가 세부적으로 이루어짐.

# [차별 점 및 배울 점]

## [차별 점]

- convNext v2 base와 Eva02 Large 모델 두 가지 사용
- 한 모델 당 Phase 1, 2로 학습을 두 번씩 거침
- T scaling 후 각 추론 후 Soft Voting을 진행

## [배울 점]

- 여러 증강(Rotating, Perspective, HalfCut, Cutmix, Mixup)을 사용하였고, 가장 성능 향상에 도움이 된 HalfCut, Cutmix, Erasing 도출
- Basemodel 를 기본 삼아 SoftVoting 실행.

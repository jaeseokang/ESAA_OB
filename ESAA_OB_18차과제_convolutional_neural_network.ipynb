# Ch5. 합성곱 신경망 1

- 딥러닝의 역전파를 보면, 순전파 과정에 따라 계산된 오차 정보가 신경망의 모든 노드로 전송됨. 이러한 과정은 복잡하고 많은 자원(CPU or GPU, 메모리)을 요구
-> 이러한 문제를 해결하고자 하는 것이 "합성곱 신경망"

### 5.1.1 합성곱층의 필요성
- 이미지를 펼쳐서 분석하면 데이터 공간적 구조가 무시하게 되는데, 이것을 방지하려고 도입된 것이 합성곱층임.

### 5.1.2 합성곱 신경망 구조(CNN)
- 5가지 층으로 구성
- 입력층, 합성곱층, 풀링층, 완전연결층, 출력층

  - 입력층: 입력 이미지 데이터가 최초로 거치게 되는 계층
  - 합성곱층: 입력 데이터에서 특성을 추출하는 역할 수행
    - 입력 데이터: W1 x H1 X D1 (가로, 세로, 채널 또는 깊이)
    - 하이퍼파라미터: 필터 개수(K), 필터 크기(F), 스트라이드(S), 패딩(P)
    - 출력 데이터: W2, H2, D2
  - 풀링층: 합성곱층과 유사하게 특성 맵의 차원을 다운 샘플링해 연산량을 감소시키고, 주요한 특성 벡터를 추출해 학습을 효과적으로 할 수 있게 함.
    - 풀링 연산에는 '최대 풀링', '평균 풀링' 두 가지 사용
    - 파라미터는 합성곱층과 동일
  - 완전연결층: 합성곱층과 풀링층이 거치면서 차원이 축소된 특성 맵은 최종적으로 완전쳔결층으로 전달.
  - 출력층: 소프트맥스 활성화 함수가 사용되는데, 입력받은 값을 0~1 사이의 값으로 출력. 각 레이블에 속할 확률 값이 출력. 이때 가장 높은 확률 값을 갖는 레이블이 최종 값으로 선정됨.

### 5.1.3 1D, 2D, 3D 합성곱
- 1D 합성곱: 필터가 시간을 축으로 좌우로만 이동할 수 있는 합성곱 -> 입력(W)과 필터(k)에 대한 출력은 W가 됨.
스크린샷 2025-11-16 오전 11.35.57.png

- 2D 합성곱: 방향 두 개로 움직이는 형태. 입력(W, H)과 필터(k, k)에 대한 출력은 (W, H)가 되며, 출력 형태는 2D 행렬이 됨.
스크린샷 2025-11-16 오전 11.36.03.png

- 3D 합성곱: 입력(W, H, L)에 대해 필터(k, k, d)를 적용하면 출력으로 (W, H, L)을 갖는 형태가 3D 합성곱임. 출력은 3D 형태, 이때 d<L 유지가 중요
스크린샷 2025-11-16 오전 11.36.10.png

- 3D 입력을 갖는 2D 합성곱: 입력이 3D 형태임에도 출력 형태가 3D가 아닌 2D 행렬을 취하는 것. Ex) LeNet-5, VGG
스크린샷 2025-11-16 오전 11.36.16.png

- 1 x 1 합성곱: 3D 형태로 입력됨. 즉, 입력(W, H, L)에 필터(1, 1, L)를 적용하면 출력은 (W, H)가 됨. Ex) GoogLeNet

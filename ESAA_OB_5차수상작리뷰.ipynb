- 비지도학습인 군집화(Clustering)을 이용해 주차장 유형별 구조적 이질성 패턴을 학습함. 산출물로 Cluster label, 각 그룹별 탄력 요금을 위한 수요 예측을 시행.

# [차별 점 및 배울 점]

## [차별 점]

- 단순히 주차 시설 확충, 단속 강화 같은 전통적 정책에 국한되지 않고, 데이터 기반 분석과 최신 통계 기법을 적용해 더욱 효율적인 주차 정책을 제시함
- 수요 예측의 경우, 단순히 평균 수요를 예측하는 것이 아니라 군집화와 분위수 회귀를 통해 맞춤형 정책 설계
- 정책 효과 평가의 경우, 정책 도입 전후 회전율 변화를 베이지안 모델을 통해 인과적으로 추정

## [배울 점]

- ML/DL의 통합적 분석을 통해 효과적인 정책 제안 마련.

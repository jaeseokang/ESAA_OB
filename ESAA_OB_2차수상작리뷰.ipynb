# [코드 리뷰]

(1) EDA
- 이상치 탐색 등.

(2) 기능 엔지니어링 및 데이터 정리

- 파생변수 생성(전략: 2023년, 2021년 대회를 참고해 유용한 파생변수 생성)
- 공휴일 변수 생성(휴일 카테고리 선정)
- 이상치 보간: 짧은 기간 발생하는 스파이크 데이터(정전 or 기기오류) 선형보간 방식으로 처리
- 이상치 제거: 긴 기간에 발생하는 스파이크, 원인을 알 수 없는 평균 이동 제거
- 전력 통계량: Validation & Test 설정, 파생 변수, Validation/Test 분리 규칙 준수 + 이상치 처리 후 통계량 활용(신뢰성 있는 전력 특성 반영)

(3) 모델 훈련 및 생성

- 활용 모델: TabPFN [1], 구조직 인과 모델 기반, 1억 개 이상의 합성 데이터셋 구축
- 경진대회 데이터에 적용: 건물별 모델링 데이터에 적합, 시계열 특성(사인 코사인 변수로 활용), 다양한 실험을 해야하는 경진대회 특성(건물 1개당 수초 내로 완료)
- 건물 별로 최적의 변수 선택: 방법론 - 후진 제거법, 건물 맞춤형 최적 변수 조합 도출
- 모델 seed ensemble: 딥러닝 모델은 초기값(seed)에 따라 결과 크게 달라짐 -> Seed Ensemble 적용(동일 모델, 동일 환경에서 Seed 만 변경), 100회 반복&테스트 진행

(4) 후처리
- 모든 경우에 대해 성능 증가 확인.

# [차별 점 및 배울 점]

## [차별 점]

- nature에 발표된 tabular data에 SOTA를 달성한 TabPFN [1]이라는 딥러닝 모델 활용
- 각 건물별 맞춤형 최적의 변수 찾아줌.

## [배울 점]

- tabular 데이터만 딥러닝이 활약하지 못하는 점을 착안해 문제 제기.
- 문제 해결 위해 구조적 인과 모델 활용해 1억개 이상의 합성 데이터 구축
- 각 건물 별로 최적의 solution 제공
